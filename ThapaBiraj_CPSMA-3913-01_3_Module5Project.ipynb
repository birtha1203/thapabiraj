# Check whether it's Tree or Graph

```python
def check(degree, n): 
      
    # Find sum of all degrees 
    deg_sum = sum(degree) 
      
    # It is tree if sum is equal to 2(n-1) 
    if (2*(n-1) == deg_sum): 
        return True
    else: 
        return False
      
#main 
n = 7
degree = [2, 3, 1, 1, 3, 1, 1]; 
if (check(degree, n)): 
    print("Tree")
else: 
    print ("Graph")
```
